# Line plot stages of fMRI cleaning



```python
import pathlib


rawdata_base_path = pathlib.Path("/mnt/rawdata/new_conte/")
derivatives_base_path = pathlib.Path("/mnt/derivatives/new_conte/")
test_sub = "sub-U04385"
test_ses = "ses-21084"
test_run = "ses-21084_task-image_run-02"

fmri_rawdata_file = test_sub + "_" + test_ses + "_task-image_run-02_bold.nii.gz"
rawdata_path = rawdata_base_path / test_sub / test_ses / "func" / fmri_rawdata_file
print("Initial file '{}' {}.".format(
    str(rawdata_path),
    "exists" if rawdata_path.exists() else "does not exist",
))

fmri_final_file = test_ses + "_task-image_run-02.nii.gz"
derivatives_path = derivatives_base_path / "abcd" / test_sub / "ses-None" / "files" / "MNINonLinear" / "Results" / test_run / fmri_final_file
print("Final file '{}' {}.".format(
    str(derivatives_path),
    "exists" if derivatives_path.exists() else "does not exist",
))
```

```python
import matplotlib.pyplot as plt
from nilearn.plotting import plot_carpet
import nibabel as nib


rawdata_image = nib.load(rawdata_path)
fig, ax = plt.subplots()
display = plot_carpet(rawdata_image, axes=ax)
```

```python
final_image = nib.load(derivatives_path)
fig, ax = plt.subplots()
display = plot_carpet(final_image, axes=ax)
```
